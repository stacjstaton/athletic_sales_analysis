```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames
athletic_sales_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020.head()
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020.dtypes
```

```python
# Check the 2021 sales data types.
athletic_sales_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_analysis = pd.concat([athletic_sales_2020, athletic_sales_2021] ,axis= "rows",join="inner")
athletic_sales_analysis
```

```python
# Check if any values are null.
athletic_sales_analysis.isnull().sum()
```

```python
# Check the data type of each column
athletic_sales_analysis.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_sales_analysis["invoice_date"] = pd.to_datetime(athletic_sales_analysis["invoice_date"])
```

```python
# Confirm that the "invoice_date" data type has been changed.
athletic_sales_analysis.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
product_sales = athletic_sales_analysis.groupby(['region', 'state', 'city'])["units_sold"].agg(Total_Products_Sold=("sum"))
# Rename the sum to "Total_Products_Sold".
product_sales.rename({"sum": "Total_Products_Sold"}, inplace=True)

# Show the top 5 results.
product_sales.sort_values(by="Total_Products_Sold", ascending = False).head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
product_sales = pd.pivot_table(athletic_sales_analysis, values= "units_sold", index=["region", "state", "city"], aggfunc="sum")
                                                                              
# Rename the "units_sold" column to "Total_Products_Sold"
product_sales.rename(columns={"units_sold": "Total_Products_Sold"}, inplace=True)

# Show the top 5 results.
product_sales.sort_values(by="Total_Products_Sold", ascending=False).head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
most_sales = athletic_sales_analysis.groupby(['region', 'state', 'city'])["total_sales"].agg(total_sales=("sum"))
# Rename the "total_sales" column to "Total Sales"
most_sales.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Show the top 5 results.
most_sales.sort_values(by="Total Sales", ascending=False).head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales = pd.pivot_table(athletic_sales_analysis, values= "total_sales", index=["region", "state", "city"], aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Show the top 5 results.
total_sales.sort_values(by="Total Sales", ascending = False).head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
women_footwear_sales_data = athletic_sales_analysis[athletic_sales_analysis["product"].str.contains("women's athletic footwear", case=False)]
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
women_footwear_sales_data = athletic_sales_analysis.groupby(['retailer','region', 'state', 'city'])["total_sales"].agg(total_sales=("sum"))
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_footwear_sales_data.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"}, inplace=True)


# Show the top 5 results.
women_footwear_sales_data.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
women_footwear_sales_data = pd.pivot_table(athletic_sales_analysis, values= "total_sales", index=["retailer", "region", "state", "city"], aggfunc="sum")

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_footwear_sales_data.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"}, inplace=True)

# Show the top 5 results.
women_footwear_sales_data.head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
women_footwear_sales_data = pd.pivot_table(athletic_sales_analysis,index= ['invoice_date'], values= "total_sales",aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
women_footwear_sales_data.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Show the table.
women_footwear_sales_data.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = women_footwear_sales_data.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
sorted_daily_sales = daily_sales.sort_values(by='Total Sales', ascending=False)
print(sorted_daily_sales)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
daily_sales = women_footwear_sales_data.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
sorted_daily_sales = daily_sales.sort_values(by='Total Sales', ascending=False)
print(sorted_daily_sales)
```
